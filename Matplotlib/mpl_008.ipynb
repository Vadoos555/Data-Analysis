```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(0, 5, 0.25), label='Line 1')
ax.plot(np.arange(0, 10, 0.5), label='Line 2')
ax.legend()
ax.grid()
```

```python
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot()
ax.plot(np.arange(0, 5, 0.25), '--o', label='Line 1')
ax.plot(np.arange(0, 10, 0.5), ':s', label='Line 2')
ax.legend()
```

```python
from matplotlib.lines import Line2D

l1 = Line2D([1, 2, 3], [1, 2, 4])

fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot()
ax.plot(np.arange(0, 5, 0.25), '--o', label='line1')
ax.plot(np.arange(0, 10, 0.5), ':s', label='line2')

ax.add_line(l1)
```

```python
from matplotlib.lines import Line2D

l1 = Line2D([1, 2, 3], [1, 2, 4])

fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot()

ax.add_line(l1)
ax.set(xlim=(1, 3), ylim=(1, 4))
```
