```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
slices = [60, 40]
labels = ['Mercedes', 'Audi']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")
```

```python
slices = [60, 40, 30, 20]
labels = ['Mercedes', 'Audi', 'Volvo', 'Opel']
colors = ['blue', 'red', 'green', 'yellow']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}, colors=colors)
plt.title("My Awesome Pie Chart")
```

```python
slices = [59219, 55466, 47544, 36443, 35917, 31991, 27097, 23030, 20524, 18523, 18017, 7920, 7331, 7201, 5833]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java', 'Bash/Shell/PowerShell', 'C#', 'PHP', 'C++', 'TypeScript', 
          'C', 'Other(s):', 'Ruby', 'Go', 'Assembly']

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'})
plt.title("My Awesome Pie Chart")
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}, explode=explode)
plt.title("My Awesome Pie Chart")
```

```python
slices = [59219, 55466, 47544, 36443, 35917]
labels = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
explode = [0, 0, 0, 0.1, 0]

plt.pie(slices, labels=labels, wedgeprops={'edgecolor': 'black'}, explode=explode, shadow=True, startangle=90, autopct='%1.1f%%')
plt.title("My Awesome Pie Chart")
```
