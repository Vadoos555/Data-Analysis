```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.set_xticklabels([])
ax.set_yticklabels([])
```

```python
from matplotlib.ticker import NullFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.xaxis.set_major_formatter(NullFormatter())
```

```python
from matplotlib.ticker import NullFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.yaxis.set_major_formatter(NullFormatter())
```

```python
from matplotlib.ticker import FormatStrFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.yaxis.set_major_formatter(FormatStrFormatter('y=%.2f'))
```

```python
from matplotlib.ticker import FuncFormatter


def format_0y(x, pos):
    return f'[{x}]' if x < 0 else f'({x})'


fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.grid()
ax.yaxis.set_major_formatter(FuncFormatter(format_0y))
```

```python
from matplotlib.ticker import ScalarFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x) * 1e5)

sf = ScalarFormatter()
sf.set_powerlimits((-4, 4))

ax.grid()
ax.yaxis.set_major_formatter(sf)
```

```python
from matplotlib.ticker import ScalarFormatter
import matplotlib

matplotlib.rcParams["axes.formatter.limits"] = (-5, 5)

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x) * 1e5)

ax.grid()
```

```python
from matplotlib.ticker import FixedFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x) * 1e5)

ax.grid()
ax.xaxis.set_major_formatter(FixedFormatter([-3, -2, -1, 0, 1, 2, 3]))
```

```python
from matplotlib.ticker import FixedFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-np.pi/2, np.pi, 0.1)
ax.plot(x, np.sin(x) * 1e4)

ax.grid()
ax.xaxis.set_major_formatter(FixedFormatter(['a', 'b', 'c', 'd', 'e', 'f', 'g']))
```
