```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

y = np.arange(0, 5, 1)
x = np.array([a*a for a in y])
y2= [0, 2, 3, 4, 5, 7]
x2 = [i+1 for i in y2]

plt.plot(x, y, x2, y2)
plt.grid()
```
