```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.random.normal(0, 2, 500)
plt.hist(y)
plt.grid()
```

```python
y = np.random.normal(0, 2, 500)
plt.hist(y, 50, label='histograma')
plt.legend(loc='upper right')
plt.grid()
```

```python
x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(1, 5, len(x))

plt.bar(x, y)
plt.grid()
```

```python
y = np.random.normal(0, 2, 500)
x = np.linspace(np.min(y), np.max(y), 10)

bars = [len(y[np.bitwise_and(y >= x[i], y < x[i+1])]) for i in range(len(x)-1)]
plt.bar(range(len(x)-1), bars)
plt.grid()
```

```python
y = np.random.normal(0, 2, 500)
x = np.linspace(np.min(y), np.max(y), 10)

bars = [len(y[np.bitwise_and(y >= x[i], y < x[i+1])]) for i in range(len(x)-1)]
plt.barh(range(len(x)-1), bars)
plt.grid()
```

```python
x = [f'H{i+1}' for i in range(10)]
y = np.random.randint(-20, 20, len(x))

plt.bar(x, y, width=0.5, linewidth=2, edgecolor='r', yerr=2, bottom=10, color='c')
plt.grid()
```

```python
x = np.arange(10)
y1 = np.random.randint(3, 20, len(x))
y2 = np.random.randint(3, 20, len(x))
w = 0.3

plt.bar(x - w/2, y1, width=w)
plt.bar(x + w/2, y2, width=w)
plt.grid()
```

```python
vals = [10, 40, 23, 30, 7]
labels = ['Toyota', 'BMW', 'Lexus', 'Audi', 'Opel']
plt.pie(vals, labels=labels)
```

```python
vals = [10, 40, 23, 30, 7]
labels = ['Toyota', 'BMW', 'Lexus', 'Audi', 'Opel']
exp = (0.1, 0.2, 0, 0, 0)
plt.pie(vals, labels=labels, autopct='%.2f', explode=exp, shadow=True)
```

```python
vals = [10, 40, 23, 30, 7]
labels = ['Toyota', 'BMW', 'Lexus', 'Audi', 'Opel']
plt.pie(vals, labels=labels, autopct='%.2f', wedgeprops=dict(width=0.5), shadow=True)
```
