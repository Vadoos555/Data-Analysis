```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
 
x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.plot(x, np.sinc(x) * np.exp( -np.abs(x/10)))
ax.grid()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
 
x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.semilogy(x, np.sinc(x) * np.exp( -np.abs(x/10)))
ax.grid()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
 
x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.plot(x, np.sinc(x) * np.exp( -np.abs(x/10)))
ax.set_yscale('log')
ax.grid()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
 
x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.plot(x, np.sinc(x) * np.exp( -np.abs(x/10)))
ax.set_yscale('log')
ax.set_xscale('log')
ax.grid()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
 
x = np.arange(-10*np.pi, 10*np.pi, 0.5)
ax.plot(x, np.sinc(x) * np.exp( -np.abs(x/10)))
ax.set_yscale('log', subs=[2, 5, 9])
ax.grid()
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
 
x = np.arange(-10*np.pi, 10*np.pi, 0.1)
ax.plot(x, np.sinc(x) * np.exp( -np.abs(x/10)))
ax.set_xscale('symlog', linthresh=2)
ax.grid()
```
