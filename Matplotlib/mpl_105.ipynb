```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data_555.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']

# overall_median = 57287

plt.plot(ages, dev_salaries, color='yellow', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='blue', label='Python')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
```

```python
data = pd.read_csv('data_555.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']

overall_median = 57287

plt.plot(ages, dev_salaries, color='yellow', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='blue', label='Python')

plt.fill_between(ages, py_salaries, overall_median, alpha=0.5)

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.grid()
```

```python
data = pd.read_csv('data_555.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']

overall_median = 57287

plt.plot(ages, dev_salaries, color='yellow', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='blue', label='Python')

plt.fill_between(ages, py_salaries, overall_median, where=( py_salaries>overall_median), interpolate=True, alpha=0.5)

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.grid()
```

```python
data = pd.read_csv('data_555.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']

overall_median = 57287

plt.plot(ages, dev_salaries, color='yellow', linestyle='--', label='All Devs')
plt.plot(ages, py_salaries, color='blue', label='Python')

plt.fill_between(ages, py_salaries, overall_median, where=( py_salaries>overall_median), interpolate=True, alpha=0.5)
plt.fill_between(ages, py_salaries, overall_median, where=( py_salaries<overall_median), interpolate=True, alpha=0.5, color='red')

plt.legend()
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.grid()
```
