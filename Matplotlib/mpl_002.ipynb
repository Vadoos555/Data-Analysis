```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([1, 2, -6, 0, 4])
plt.plot(y)
```

```python
x = np.array([4, 5, 6, 7, 8])
plt.plot(x, y)
```

```python
x = np.array([1, 1, 5, 5, 1])
y = np.array([1, 5, 5, 1, 1])
plt.plot(x, y)
```

```python
y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]

plt.plot(x, y, x2, y2)
plt.grid()
```

```python
y = np.arange(0, 5)
x = np.array([a*a for a in y])
plt.plot(x, y, '--')
plt.grid()
```

```python
y = np.arange(0, 5)
x = np.array([a*a for a in y])
y2 = [0, 1, 2, 3]
x2 = [i+1 for i in y2]

plt.plot(x, y, '--bo', markerfacecolor='w')
plt.plot(x2, y2, '-.rs', markerfacecolor='w')
plt.grid()
```

```python
lines = plt.plot(x, y, '--go', markerfacecolor='w')
plt.setp(lines[0], linestyle='-.', marker='s', markerfacecolor='b', linewidth=4)

plt.plot(x2, y2, '-.rs', markerfacecolor='w')
plt.grid()
```

```python
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.cos(x)
plt.plot(x, y)
plt.fill_between(x, y)
plt.grid()
```

```python
plt.plot(x, y)
plt.fill_between(x, y, where=(y > 0),))
plt.grid()
```

```python
plt.plot(x, y)
plt.grid()
plt.fill_between(x, y, where=(y < 0), color='r', alpha=0.5)
plt.fill_between(x, y, where=(y > 0), color='g', alpha=0.5)
```
