```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot()

x = np.arange(0, 10)
ax.step(x, x)
ax.grid()
```

```python
fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot()

x = np.arange(0, 10)
ax.step(x, x, '-bo')
ax.grid()
```

```python
x = np.arange(0, 10)
plt.step(x, x, '-go', where='post')
plt.grid()
```

```python
x = np.arange(0, 10)
plt.step(x, x, '-rs', where='mid')
plt.grid()
```

```python
x = np.arange(0, 10)
plt.step(x, x, '-bo', x, np.cos(x), '--rx', where='mid')
plt.grid()
```

```python
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot()

x = np.arange(-2, 2, 0.1)
y1 = np.array([-y**2 for y in x]) + 8
y2 = np.array([-y**2 for y in x]) + 8
y3 = np.array([-y**2 for y in x]) + 8

ax.stackplot(x, y1, y2, y3)
ax.grid()
```

```python
x = np.arange(-2, 2, 0.1)
y1 = np.array([-y**2 for y in x]) + 8
y2 = np.array([-y**2 for y in x]) + 8
y3 = np.array([-y**2 for y in x]) + 8

plt.stackplot(x, y1, y2, y3)
plt.legend(('line 1', 'line 2', 'line 3'))
plt.grid()
```

```python
x = np.arange(-np.pi, np.pi, 0.3)
plt.stem(x, np.cos(x))
plt.grid()
```

```python
x = np.arange(-np.pi, np.pi, 0.3)
plt.stem(x, np.cos(x), '--g', bottom=0.5, label='cosinus')
plt.legend()
plt.grid()
```

```python
x = np.arange(-np.pi, np.pi, 0.3)
plt.stem(x, np.cos(x), '--b', bottom=0.5, label='cosinus')
plt.legend()
plt.grid()
```
