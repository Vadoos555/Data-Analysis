```python
fig, ax = plt.subplots(2, 2)
fig.set_size_inches(7, 4)
fig.set_facecolor('#eee')

ax[0, 0].plot(np.arange(0, 5, 0.2))
ax[0, 0].grid()
ax[0, 1].plot(np.arange(5, 0, -0.2))
ax[0, 1].grid()
```
