```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Bob', 'Jane', 'John', 'Vadim', np.nan, None, 'NA'], 
    'last': ['Schaher', 'Doe', 'Doe', 'Smart', np.nan, np.nan, 'Missing'], 
    'email': ['BobSchaher@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df.dropna()
```

```python
df.dropna(how='all')
```

```python
df.dropna(axis=1, how='all')
```

```python
df.dropna(subset=['email'])
```

```python
df.dropna(how='all', subset=['last', 'email'])
```
